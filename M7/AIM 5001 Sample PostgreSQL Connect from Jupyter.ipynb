```python
import pandas as pd
import psycopg2
import sqlalchemy
import matplotlib as plt
```

```python
###################################################################
# Make sure you edit the host, database, user, and password settings
# as appropriate for your local PostgreSQL environment !!!
###################################################################

conn = psycopg2.connect(
    host="localhost",
    database="chinook",
    user="postgres",
    password="postgres")
```

```python
# Create a cursor object
cur = conn.cursor()

# A sample query of a count of all data from the "Artist" table in the "Chinook" database
cur.execute("""SELECT COUNT(*) FROM "Artist" """)
query_results = cur.fetchall()
print(query_results)
```

```python
# A sample query of a count of all data from the "Playlist" table in the "Chinook" database
cur.execute("""SELECT COUNT(*) FROM "Playlist" """)
query_results = cur.fetchall()
print(query_results)
```

```python
# A sample query of a count of all data from the "Track" table in the "Chinook" database
cur.execute("""SELECT COUNT(*) FROM "Track" """)
query_results = cur.fetchall()
print(query_results)
```

```python
# create a Pandas dataframe using the content of the "Album" table
df = pd.read_sql_query("""SELECT * FROM "Album" """, conn)
```
